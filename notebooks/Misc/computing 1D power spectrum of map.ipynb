### Load map (and display to check it's good)

### Compute its power spectrum (and display to check it's good)

```python
data_psd = mrcutils.data2psd(data, log=False)
```

### Compute 1D PSD from 3D PSD

```python
psd_1d = mrcutils.data2radialprofile(data_psd)
```
